```python
def split_cpu(x):
    x = x.split(" ")
    last_part = x[-1]  # Get the last part of the string
    # Extract numeric part from the last part
    numeric_part = ''.join(filter(str.isdigit, last_part))
    
    if numeric_part:
        gen_number = int(numeric_part[0])  # Use the first digit as the generation number
        # Map the generation number to the format
        if gen_number == 1:
            return "1st gen"
        elif gen_number == 2:
            return "2nd gen"
        elif gen_number == 3:
            return "3rd gen"
        else:
            return f"{gen_number}th gen"
    else:
        return "Unknown gen"

df['Cpu Gen'] = df['Cpu Model'].apply(lambda x: split_cpu(x))
df.head(20)
```

### Memory Slicing

```python
def split_str(x):

    x = x.split(" ")
    if len(x)<=3:
        r = x[0]
    elif len(x)>3:
        r = f"{x[0]} + {x[-2]}" 
    return r

df['Storage'] = df['Memory'].apply(lambda x : split_str(x))
df.head()
```

```python
def extract_storage(memory_string, memory_type):
    # avoid mismatch
    memory_string = memory_string.lower()
    parts = memory_string.split('+')
    
    for part in parts:
        # remove whitespaces
        part = part.strip()
        if memory_type in part:
            # get numeric value before memory type
            storage_value = ''.join(filter(str.isdigit, part))
            return int(storage_value)
    
    return 0
    # mapping
def preprocess_memory_data(df):
    memory_types = {
        'SSD': 'ssd',
        'HDD': 'hdd',
        'Flash Storage': 'flash storage',
        'Hybrid': 'hybrid'
    }
    
    for memory_type, keyword in memory_types.items():
        df[memory_type] = df['Memory'].apply(lambda x: extract_storage(x, keyword))
    
    return df


df = preprocess_memory_data(df)
df
```

# Feature & Model Selection

```python
columns_to_select = [0,1,2,3,4,5,6] + list(range(-9, 0))
x=df.iloc[:, columns_to_select]
y=df.iloc[:,6]
```

# Model Evaluation

```python
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

### SGDRegressor

```python
model = SGDRegressor(max_iter=1000)
y_train = y_train.values.reshape(-1, 1)
y_test = y_test.values.reshape(-1, 1)
model.fit(x_train, y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```
